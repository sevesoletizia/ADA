```python
import pandas as pd
import os

PATH = './data/'
file_name = os.path.join(PATH,'interactive_data.csv')
df = pd.read_csv(file_name, index_col=0)
```

```python
# Removing the None selected
df_fil = df.loc[~(df == 'None selected').any(axis=1)]
df_fil.head(5)
```

```python
tot = df_fil['Deaths'].sum()
print(f'Total gun deaths: {tot}')

suicides = df_fil[df_fil['Intent'] == 'Suicide']['Deaths'].sum()
print(f'Suicides: {suicides}')

print(f'{(suicides/tot)*100:.2f}% of gun deaths are suicides')
```

```python
suicides_male = df_fil[(df_fil['Intent'] == 'Suicide') 
                       & (df_fil['Gender'] == 'Male')]['Deaths'].sum()
print(f'Suicides commited by males: {suicides_male}')

print(f'{(suicides_male/suicides)*100:.2f}% of suicides victims are male')
```

```python

homicides = df_fil[df_fil['Intent'] == 'Homicide']['Deaths'].sum()
print(f'Homicides: {homicides}')

print(f'{(homicides/tot)*100:.2f}% of gun deaths are homicides')
```

```python
homicides_male_15_34 = df_fil[(df_fil['Intent'] == 'Homicide') 
                        & (df_fil['Gender'] == 'Male') 
                        & (df_fil['Age'] == '15 - 34')]['Deaths'].sum()

print(f'Homicide victims males in age 15-34: {homicides_male_15_34}')


homicides_male_15_34_black = df_fil[(df_fil['Intent'] == 'Homicide') 
                        & (df_fil['Gender'] == 'Male') 
                        & (df_fil['Age'] == '15 - 34')
                        & (df_fil['Race'] == 'Black')]['Deaths'].sum()
print(f'Homicide victims by  black males in age 15-34: {homicides_male_15_34_black}')


print(f'{(homicides_male_15_34_black/homicides_male_15_34)*100:.2f}% of homicide victims who are males in the age-group of 15--34 are black')
```

```python
homicides_female = df_fil[(df_fil['Intent'] == 'Homicide') 
                       & (df_fil['Gender'] == 'Female')]['Deaths'].sum()
print(f'Homicide victims female: {homicides_female}')

print(f'{(homicides_female/homicides)*100:.2f}% of homicide victims are female')
```
